```python
import string
from nltk.corpus import stopwords

def text_cleaning(text):
    text_punc_removed = [char for char in text if char not in string.punctuation]
    text_punc_removed_joined = ''.join(text_punc_removed)
    text_punc_removed_joined_cleaned = [char for char in text_punc_removed_joined.split() if char.lower() not in stopwords.words('english')]
    
    return text_punc_removed_joined_cleaned
```

```python
imp_keys = []
for i in imp_keywords:
    for word in i.split():
        imp_keys.append(word)
        
imp_keys
```

```python
from collections import Counter

imp_key_collection = Counter(imp_keys)
imp_key_collection
```

```python
imp_key_list = imp_key_collection.keys()
imp_key_list
```

```python
def text_cleaning_refined(text):
    text_punc_removed = [char for char in text if char not in string.punctuation]
    text_punc_removed_joined = ''.join(text_punc_removed)
    text_punc_removed_joined_cleaned = [char for char in text_punc_removed_joined.split() if char.lower() not in stopwords.words('english')]
    text_imp = [char for char in text_punc_removed_joined_cleaned if char.lower() in imp_key_list]
    
    return text_imp
```

# From CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

vector = CountVectorizer(analyzer=text_cleaning)
df_cntvect = vector.fit_transform(df['complaint'])
```

```python
print(vector.get_feature_names_out())
```

```python
print(df_cntvect.toarray())
```

```python
df_cntvect.shape
```

```python
label = df['IPC_Section'].values
label
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_cntvect, label, test_size=0.2, shuffle=True)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_class = MultinomialNB()
NB_class.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = NB_class.predict(x_test)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred))
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, cmap='Greens', annot=True)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)
```

# From TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer

text_tfidf = TfidfTransformer().fit_transform(df_cntvect)
```

```python
x_train, x_test, y_train, y_test = train_test_split(text_tfidf, label, test_size=0.2, shuffle=True)

NB_class = MultinomialNB()
NB_class.fit(x_train, y_train)

y_pred = NB_class.predict(x_test)

print(classification_report(y_test, y_pred))
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, cmap='Greens', annot=True)
accuracy_score(y_pred, y_test)
```
