### Movie reccomender for entire data set

```python
user_title_mat
```

```python
movie_corr = user_title_mat.corr(method = 'pearson', min_periods=80)
```

```python
movie_corr.head(10)
```

```python
my_ratings_df = pd.read_csv('../Data Science/Udemy_Data/My_Ratings.csv')
my_ratings_df
```

```python
my_ratings_df['Movie Name'][0]
```

```python
similar_movies_list = pd.Series()
length = my_ratings_df.shape[1]

for i in range(0, length):
    similar_movies = movie_corr[my_ratings_df['Movie Name'][i]].dropna()
    similar_movies = similar_movies.map(lambda x: x*my_ratings_df['Ratings'][i])
    similar_movies_list = similar_movies_list._append(similar_movies)
```

```python
similar_movies_list.sort_values(inplace=True, ascending=False)
print(similar_movies_list.head(10))
```
